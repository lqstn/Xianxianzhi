```python
#数据库导入
import matplotlib.pyplot as plt
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Geo,Map,Timeline

plt.rcParams["font.sans-serif"]=["Microsoft YaHei"]
plt.rcParams["axes.unicode_minus"]=False
```

```python
reader = pd.read_excel("events.xlsx")
reader.to_csv("events.csv",index=False,header=False)

df = pd.read_csv("events.csv").dropna().drop(["no"],axis=1)
df
```

```python
#中国地图 删除国外的地点
#目前只有巴黎、莫斯科
df = df.drop(df[df['location'].isin(['巴黎', '莫斯科'])].index).reset_index().drop(["index"],axis=1)
df
```

```python
# Timeline
# 生成时间轴
tl = Timeline(init_opts=opts.InitOpts(width="1425px",height="760px",bg_color="#bd925a"))

# 设置时间轴的属性(autoplay,播放速度,标签色号)
tl.add_schema(pos_bottom=30,is_auto_play=True, play_interval=2100,linestyle_opts=opts.LineStyleOpts(color="#670001"),itemstyle_opts=opts.ItemStyleOpts(color ="#670001"),label_opts = opts.LabelOpts(color ="#94070A"))
# 生成地图
# for i in range(1949,2021):
# drop df["year"] time

#show only the year in df["year"] at the map's timeline, now it shows all the years and date and time



for i in year_loc_evtcnt["year"].unique():
    map = (
        Geo(init_opts=opts.InitOpts(width="1425px",height="760px",bg_color="#f0ebe5"))
        .add_schema(maptype="china",itemstyle_opts=opts.ItemStyleOpts(color="#fffaf4",border_color="#94070A"))
        .set_global_opts(title_opts=opts.TitleOpts(title="中国共产党大事记",title_textstyle_opts=opts.TextStyleOpts(color="#670001",font_size=30),pos_left="left",
                                                   subtitle=str(i)+"年",subtitle_textstyle_opts=opts.TextStyleOpts(color="#94070A",font_size=22)),
                                                    legend_opts=opts.LegendOpts(is_show=True,textstyle_opts=opts.TextStyleOpts(color="#670001"))))
    # tooltip_opts=opts.TooltipOpts(is_show=True,formatter="曾发生过{}个大事记")

    year_loc_evtcnt =  df.groupby(["year","location"]).size().reset_index()
    year_loc_evtcnt.columns = ["year","location","countt"]

    # load data进Geo Map
    map.add("事件发生地点",[z for z in zip(year_loc_evtcnt[year_loc_evtcnt["year"]==i].location, year_loc_evtcnt[year_loc_evtcnt["year"]==i].countt)],symbol_size=35,type_="effectScatter",color="#670001")
    map.set_series_opts(label_opts=opts.LabelOpts(is_show=True, formatter="{b}",color="#fffaf4"))
    tl.add(map, f"{i}年")
tl.render("timeline2.html")
```
